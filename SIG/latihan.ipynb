```python
import pandas as pd
import geopandas as gpd
import folium
```

```python
lat = -5.5111966
long =122.871147
```

```python
m = folium.Map(
    location=[lat, long],
    zoom_start=16
)
```

```python
folium.Marker(
    location=[-5.511966, 482.871147],
    tooltip="Lokasi Wahyu"
).add_to(m)
```

```python
folium.Marker(
    location=[-5.512300, 122.872000],
    tooltip="Lokasi Zaliyanti"
).add_to(m)
```

```python
folium.Polygon(
    locations=[
        [-5.511966, 122.871147],
        [-5.511700, 122.871700],
        [-5.512300, 122.872000],
        [-5.512600, 122.871300]
    ],
    color="red",
    fill=True,
    fill_color="red",
    fill_opacity=0.4,
    tooltip="Area Rumah Mutiyah"
).add_to(m)
```

```python
folium.PolyLine(
    locations=[
        [-5.511966, 122.871147],
        [-5.511850, 122.871220],
        [-5.511720, 122.871350],
        [-5.511600, 122.871500],
        [-5.511480, 122.871650],
        [-5.511350, 122.871800]
    ],
    color="blue",
    weight=5,
    tooltip="Jalan Poros"
).add_to(m)
```

```python
m.save('map.html')
```
